```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions and gradient descent parameters
def f_1d(x):
    return x**2  # 1D quadratic function

def grad_f_1d(x):
    return 2 * x  # Gradient of the 1D function

def gradient_descent_1d(x0, step_size, iterations):
    x_vals = [x0]
    for _ in range(iterations):
        x_new = x_vals[-1] - step_size * grad_f_1d(x_vals[-1])
        x_vals.append(x_new)
    return x_vals

# Parameters for 1D case
x0_1d = 4  # Starting point
step_size_1d = 0.1
iterations_1d = 10
x_vals_1d = gradient_descent_1d(x0_1d, step_size_1d, iterations_1d)
f_vals_1d = [f_1d(x) for x in x_vals_1d]

# 1D Plot
x_plot = np.linspace(-5, 5, 100)
f_plot = f_1d(x_plot)

plt.figure(figsize=(12, 5))

# Plot 1D function and gradient descent steps
plt.subplot(1, 2, 1)
plt.plot(x_plot, f_plot, label="$f(x) = x^2$")
plt.plot(x_vals_1d, f_vals_1d, 'o-', label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent in 1D (f(x) = x^2)")
plt.legend()
plt.grid(True)

# 2D case
def f_2d(x, y):
    return x**2 + y**2  # 2D quadratic function

def grad_f_2d(x, y):
    return np.array([2 * x, 2 * y])  # Gradient of the 2D function

def gradient_descent_2d(x0, y0, step_size, iterations):
    x_vals, y_vals = [x0], [y0]
    for _ in range(iterations):
        grad = grad_f_2d(x_vals[-1], y_vals[-1])
        x_new = x_vals[-1] - step_size * grad[0]
        y_new = y_vals[-1] - step_size * grad[1]
        x_vals.append(x_new)
        y_vals.append(y_new)
    return x_vals, y_vals

# Parameters for 2D case
x0_2d, y0_2d = 3, 4  # Starting point
step_size_2d = 0.1
iterations_2d = 15
x_vals_2d, y_vals_2d = gradient_descent_2d(x0_2d, y0_2d, step_size_2d, iterations_2d)

# 2D Plot
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f_2d(X, Y)

plt.subplot(1, 2, 2)
contour = plt.contour(X, Y, Z, levels=20, cmap="viridis")
plt.plot(x_vals_2d, y_vals_2d, 'o-', color="red", label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent in 2D (f(x, y) = x^2 + y^2)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
